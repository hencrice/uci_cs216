### Calculate the endpoints of each line segment that define the edge of book spines

Compute the covariance matrix for points that belong to each connected component, then obtain the eigenvector correspond to the largest eigenvalue as the principle component

```python
locRowCol=transpose(nonzero(gradDirBinCC==unique(gradDirBinCC)[1]))*1.0
y, x=locRowCol[:,0], locRowCol[:,1]
yMean, xMean=mean(y), mean(x)
yTilde, xTilde=y-yMean, x-xMean
sumXY=sum(xTilde*yTilde)
D=((sum(xTilde**2), sumXY),(sumXY, sum(yTilde**2)))
eigVal, eigVec=eig(D)
largestEigVec=eigVec[:,argmax(eigVal)] # largestEigVec (x, y)
# compute the dot product of points in this connected component with the principle component to extract points
# that are at the extreme, then we will know the starting and ending point (denote as pt1, pt2) of this line segment
xyDotProd=dot(hstack((xTilde[:,None],yTilde[:,None])), largestEigVec)
x1,y1=hstack((x[:,None], y[:,None]))[argmin(xyDotProd)]
x2,y2=hstack((x[:,None], y[:,None]))[argmax(xyDotProd)]
angle=arctan2(largestEigVec[1],largestEigVec[0])
# lineMidPt=xMean*cos(angle)+yMean*sin(angle)
```

Use the idea above for every connected component

```python
lineParam=[]
for c in unique(gradDirBinCC):
    locRowCol=transpose(nonzero(gradDirBinCC==c))*1.0
    y, x=locRowCol[:,0], locRowCol[:,1]
    yMean, xMean=mean(y), mean(x)
    yTilde, xTilde=y-yMean, x-xMean
    sumXY=sum(xTilde*yTilde)
    D=((sum(xTilde**2), sumXY),(sumXY, sum(yTilde**2)))
    eigVal, eigVec=eig(D)
    largestEigVec=eigVec[:,argmax(eigVal)] # largestEigVec (x, y)
    # compute the dot product of points in this connected component with the principle component to extract points
    # that are at the extreme, then we will know the starting and ending point (denote as pt1, pt2) of this line segment
    xyDotProd=dot(hstack((xTilde[:,None],yTilde[:,None])), largestEigVec)
    x1,y1=hstack((x[:,None], y[:,None]))[argmin(xyDotProd)]
    x2,y2=hstack((x[:,None], y[:,None]))[argmax(xyDotProd)]
    angle=arctan2(largestEigVec[1],largestEigVec[0])
    lineParam.append((angle, x1, y1, x2, y2)) # angle, startPt, endPt of the line segment
```

Combine all the procedures above into a function:

```python
from scipy.misc import imread
from scipy.linalg import eig
from skimage.filter import canny # threshold_otsu, threshold_adaptive
from skimage.transform import hough_line, hough_line_peaks 
from skimage.morphology import remove_small_objects
from skimage.measure import label
from matplotlib.cm import gray

def bookSpineEdges(imgArr, cannySigma=5, minEdgeSize=40, thresholdFactor=0.1, deltaRad=pi/12):
    '''
    Inputs:
        imgArr: A grayscale image of book spines.
        
        cannySigma: The sigma parameter for the initial canny edge detection.
        
        minEdgeSize: After running the canny detector, edges with length below this value will be removed.
        
        thresholdFactor: A factor that will be multiplied with amax(H), where H is the Hough transform accumulator, 
            to set the minimum threshold for peaks discovery after House Transform. That is, threhold=thresholdFactor*amax(H).
        
        deltaRad: At the final filtering stage, edges within the range prominant angle+-deltaRad will be preserved.
    Output:
        An array in which each row is a line segment in the format (angle, x1, y1, x2, y2)
    '''
    # perform canny edge detection, remove edges that are too short, then compute the prominant orientation of the book spine
    cannyEdg=canny(imgArr,cannySigma)
    smallEdgeRemoved=remove_small_objects(cannyEdg, min_size=minEdgeSize)
    H,theta,distances=hough_line(smallEdgeRemoved)
    H,theta,distances=hough_line_peaks(H,theta,distances, threshold=thresholdFactor*amax(H))
    count, rad=histogram(theta,25)
    promRad=pi/2-rad[argmax(convolve(count, array([1, 1, 1], float32), 'same'))] # "pi/2-" is used to transform the original radian in rad arr to the angle system in which pos x axis is 0, rad increases counter-clockwise
    
    # compute the gradient of each pixel, group them into 16 bins, run connected component algorithm (i.e. label() function),
    # remove CC that are too small (in the paper, it's 5% of the size of the image)
    dy,dx=gradient(imgArr)
    gradDir=arctan2(dy,dx)
    gradDirBin=digitize(gradDir.flatten(), linspace(-pi,pi,16)) # 16 bins as described in the paper
    gradDirBin=reshape(gradDirBin, shape(gradDir))
    # http://scikit-image.org/docs/dev/api/skimage.measure.html?highlight=measure.label#skimage.measure.label
    gradDirBinCC=label(gradDirBin) # group points belonging to the same connected component as the same label
    gradDirBinCC=remove_small_objects(gradDirBinCC, int(round(min(shape(gradDirBinCC)))*0.05)) # remove CC that are too small (i.e. 5% of img size based on the paper)
    
    # for each connected component (C.C.), compute the covariance matrix for points belonging to that C.C.,
    # obtain the principle component (P.C.) by eig(), compute the dot product between points of this C.C. and P.C., pick 2 points
    # with extreme values as the endpoints of the line segment.
    lineParam=[]
    for c in unique(gradDirBinCC):
        locRowCol=transpose(nonzero(gradDirBinCC==c))*1.0
        y, x=locRowCol[:,0], locRowCol[:,1]
        yMean, xMean=mean(y), mean(x)
        yTilde, xTilde=y-yMean, x-xMean
        sumXY=sum(xTilde*yTilde)
        D=((sum(xTilde**2), sumXY),(sumXY, sum(yTilde**2)))
        eigVal, eigVec=eig(D)
        largestEigVec=eigVec[:,argmax(eigVal)] # largestEigVec (x, y)
        xyDotProd=dot(hstack((xTilde[:,None],yTilde[:,None])), largestEigVec)
        x1,y1=hstack((x[:,None], y[:,None]))[argmin(xyDotProd)]
        x2,y2=hstack((x[:,None], y[:,None]))[argmax(xyDotProd)]
        angle=arctan2(largestEigVec[1],largestEigVec[0])
        lineParam.append((angle, x1, y1, x2, y2)) # angle, startPt, endPt of the line segment
    
    # filter out those line segments that are either in the different orientation than the book spines or are too short/long
    lineParamArr=array(lineParam, dtype=float32)
    row,col=shape(imgArr)
    angleOk=logical_and(abs(lineParamArr[:,0])<=(promRad+deltaRad), abs(lineParamArr[:,0])>=(promRad-deltaRad)) # use abs to flip the angle to 1, 2 quadrants
    filteredArr=lineParamArr[angleOk]
    segmentLen=((filteredArr[:,1]-filteredArr[:,3])**2+(filteredArr[:,2]-filteredArr[:,4])**2)**0.5
    segmentLenOk=logical_and(segmentLen<row,segmentLen>row/4)
    return filteredArr[segmentLenOk]
```

Combine the idea in this section into a function:

```python
def mergeSegments(segments, minSpineXDistParam=0.3):
    '''
    Here I assume the thickness of books in the image won't vary too much (i.e. mean thickness*minSpineXDistParam>minimum book thickness)
    '''
    # sort the segments based on the x coordinates of each segment's 2 endpoints
    segments=segments[lexsort((segments[:,3], segments[:,1]))]
    # x distances between 2 neighboring book spine segments
    distNeighbors=abs(segments[1:,(1,3)]-segments[:-1,(1,3)])
    edgToMergeIndex=arange(len(segments))
    edgToMergeIndex=hstack((edgToMergeIndex[:len(segments)-1][:,None], edgToMergeIndex[1:][:,None]))
    edgToMergeIndex=edgToMergeIndex[any(distNeighbors<minSpineXDistParam*mean(distNeighbors),1)]
    for pair in edgToMergeIndex:
        edgToMerge=segments[pair]
        # save merged edge
        segments[pair[0]]=[mean(edgToMerge[:,0]), amax(edgToMerge[:,(1,3)]), amax(edgToMerge[:,(2,4)]), amin(edgToMerge[:,(1,3)]), amin(edgToMerge[:,(2,4)])]
    # delete the second edge in each pair of nearby book spine segments since they are already merged
    return delete(segments, edgToMergeIndex[:,1], 0)
```

### Produce quadraliteral regions that bound each book spine.

Based on the idea proposed in this thesis: http://www1.idc.ac.il/toky/msc/Thesis/Lior_talker_2013.pdf. Generate "seeds" for book spine expansion, use the midpoints of any of the 2 midpoints of 2 nearby book spine line segments obtained above. 

```python
def computeSeeds(segments):
    segIndex=arange(len(segments))
    segIndex=hstack((segIndex[:len(segments)-1][:,None], segIndex[1:][:,None]))
    seeds=[]
    for s1, s2 in segIndex:
        seg1=segments[s1]
        mid1=(seg1[1:3]+seg1[3:])/2
        seg2=segments[s2]
        mid2=(seg2[1:3]+seg2[3:])/2
        seeds.append(insert((mid1+mid2)/2, 0,(seg1[0]+seg2[0])/2))
    return array(seeds) # (x, y)
```

Remember that edge formed by (endPt1, endPt2) corresponds to v2, (endPt2, endPt3) -> to v3, (endPt3, endPt4) -> v4, (endPt4, endPt1) -> v1. Additionally, implement the edge energy function defined in the paper:

```python
def edgeEnergy(seg, edgNorm, imgGrad):
    '''
    This function implement the edge energy function defined in the paper.
    
    seg: a line segment defined by skimage.draw.line representing the edge we want to compute the energy on.
    imgGrad: A 3D Gradient array, with the 1st array in the 3rd dimension (i.e. depth) be df/dy, and the 2nd array is df/dx.
        In other words, the gradient at a location (y,x) can be accessed using imgGrad[y,x,:].
    '''
    pxlOnEdgY, pxlOnEdgX=seg # y1, x1, y2, x2
    #a.b=|a||b|cos(theta), so theta=arccos(a.b/(|a||b|))
#     print('sss: {0}'.format(norm(imgGrad[pxlOnEdgY, pxlOnEdgX, :], axis=1)))
    temp=sum(imgGrad[pxlOnEdgY, pxlOnEdgX, :]*edgNorm, 1)/norm(imgGrad[pxlOnEdgY, pxlOnEdgX, :], axis=1)
    temp[temp>1]=1
    temp[temp<-1]=-1
    return sum(arccos(temp))/len(pxlOnEdgY)
```

```python
sigma, initExpFactor, beta, lda, engyThres=1, 10, 0.8, 0.5, 2 # some hyper parameters for the model defined on p10 of the thesis

# code directly copied from previous cells just to initialize some stuffs
s=seeds[0]
# 4 unit vectors specifies 4 directions for expansion
edgNorm4=array(((cos(s[0]),sin(s[0])), (cos(s[0]+pi/2),sin(s[0]+pi/2)), (-cos(s[0]),-sin(s[0])), (-cos(s[0]+pi/2),-sin(s[0]+pi/2))), dtype=float32)
s=s[1:] # get rid of the angle part cause it's no longer needed
activeState=array([True, True, True, True]) # whether expansion in each direction is still active
expFactors=array([initExpFactor]*4,dtype=float32) # In each direction, how many unit vectors do we expand so far
# the amount of expansion in each direction might be different since expFactors might not all be the same
expVect4=(activeState*expFactors)[:,None]*edgNorm4
# s + the linear combination of any contiguous 2 rows in expVect4 will result in 4 points representing the 4 corners of our current edges
endPts4=around(s+expVect4+vstack((expVect4[1:], expVect4[0]))).astype(int16)
engyBef=[]
endPts4Bef=endPts4
# code directly copied from previous cells just to initialize some stuffs

from skimage.draw import line
from skimage.filter import gaussian_filter
dy,dx=gradient(gaussian_filter(imgs[0]/amax(imgs[0]), sigma))
grad=dstack((dx,dy)) # you can access the gradient (dx, dy) at location (a,b) by grad[a,b,:]
# for one of the direction that is active, in actual implementation this should be "for i in arange(4)[activeState]:"
for i in arange(4)[activeState]:
    x1, y1=endPts4[i]
    x2, y2=endPts4[(i+1)%4]
    mainEdg=line(y1, x1, y2, x2) # y1, x1, y2, x2
    engyBef.append(edgeEnergy(mainEdg, edgNorm4[(i+1)%4], grad))
    
for k in xrange(200):
    expFactors[activeState]+=1 # increase the total amount of expansion for those directions that are active
    expVect4=(activeState*expFactors)[:,None]*edgNorm4
    endPts4=around(s+expVect4+vstack((expVect4[1:], expVect4[0]))).astype(int16)

    # construct the edge in next iteration:
    expFactors[activeState]+=1
    expVect4=(activeState*expFactors)[:,None]*edgNorm4
    endPts4Next=around(s+expVect4+vstack((expVect4[1:], expVect4[0]))).astype(int16)
    expFactors[activeState]-=1

    # expand in active directions
    for i in arange(4)[activeState]:
        inactCond=[]
        x1, y1=endPts4[i]
        x2, y2=endPts4[(i+1)%4]
        mainEdg=line(y1, x1, y2, x2) # y1, x1, y2, x2
#         print((x2-x1, y2-y1, edgNorm4[(i+1)%4]))
        engyCurrent=edgeEnergy(mainEdg, edgNorm4[(i+1)%4], grad)

        x1Next, y1Next=endPts4Next[i]
        x2Next, y2Next=endPts4Next[(i+1)%4]
        mainEdg=line(y1Next, x1Next, y2Next, x2Next) # y1, x1, y2, x2
        engyAfter=edgeEnergy(mainEdg, edgNorm4[(i+1)%4], grad)
        
        # 3 conditions for an edge to enter inactive state
        # 1. The edge energy function is at a local min as a function of norm vector of this edge.
        #    That is, if you move the edge a little bit along its edge norm or -(edge norm), it will only result in larger energy
        inactCond.append((engyBef[i]>engyCurrent and engyAfter>engyCurrent))
#         print((i, 'engyBef[i]',engyBef[i], 'engyCurrent',engyCurrent, 'engyAfter',engyAfter, 'engyThres',engyThres))
#         engyBef[i]=engyCurrent

        # 2. The energy of this edge < predefined threshold
        inactCond.append(engyCurrent < engyThres)

        # 3. The support of the adjacent 2 edges to increment in the direction of the edge norm. This support is defined as:
        #    the maximum value of the 2 energy values of the 2 edge extension segments. Please see the figure above.
        x1Clock, y1Clock=endPts4Bef[i] # adjacent endpoint in the clockwise direction.
        x2CounterClock, y2CounterClock=endPts4Bef[(i+1)%4] # adjacent endpoint in the counter-clockwise direction
        adjExtClock, adjExtCounterClock=line(y1Clock, x1Clock, y1, x1), line(y2CounterClock, x2CounterClock, y2, x2)
        inactCond.append(max(edgeEnergy(adjExtClock, edgNorm4[i], grad), edgeEnergy(adjExtCounterClock, edgNorm4[(i+2)%4], grad))>beta*engyThres)
#         print(('adjExtClock', edgeEnergy(adjExtClock, edgNorm4[i], grad), 'adjExtCounterClock', edgeEnergy(adjExtCounterClock, edgNorm4[(i+2)%4], grad), beta*engyThres))
        activeState[i]=not all(inactCond)
#         print((i, 'inactCond', inactCond))
#     print((k, 'activeState', activeState))        
    if(sum(activeState)==0):
#         print('all inactive: {0}'.format(k))
        break

    # each inactive expansion direction may return to the active state because the edge energy it belongs to changes
    for i in arange(4)[logical_not(activeState)]:
        x1, y1=endPts4[i]
        x2, y2=endPts4[(i+1)%4]
        mainEdg=line(y1, x1, y2, x2) # y1, x1, y2, x2
        activeState[i]=(edgeEnergy(mainEdg, edgNorm4[(i+1)%4], grad)>lda*engyThres) # on p10, the description is wrong, this should be > instead of <
#         print()
    endPts4Bef=endPts4
```
